```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

DATA PREPROCESSING

```python
form=pd.read_csv('datasets/Angka Kelahiran Hasil Long Form SP2020 Menurut Kelompok Umur Ibu (Age Spesific Fertility Rate_ASFR) da.csv')
total=pd.read_csv('datasets/Angka Kelahiran Total _ Total Fertility Rate (TFR) Hasil Long Form (LF) SP2020 Menurut Provinsi_Kabupa.csv')

df=pd.merge(form, total, on='Provinsi/Kabupaten/Kota')
df
```

```python
drop_column=['idx','id','Unnamed: 3']
df=df.drop(columns=drop_column)
```

```python
df
```

```python
df_indexed=df.set_index('Provinsi/Kabupaten/Kota')
df_indexed
```

STATISTICS

```python
df_indexed.describe().round(3)
```

```python
df_mean=df_indexed.mean()
df_mean
```

VISUALIZATION

```python
from matplotlib import ticker

plt.figure(figsize=(17,9))
plt.plot(df_mean)
def percentage(x, pos):
    return '{:.0f}%'.format(x)

plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(percentage))
plt.title("Nilai rata-rata Angka kelahiran")
```

```python
plt.hist(df_indexed['25-29'], color='skyblue', edgecolor='black')

plt.ylabel('Frequency')
plt.title('Kelompok ibu melahirkan usia 25-29 ')

# Formatting x-axis ticks to include '%' symbol
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x}%'))

plt.grid(True)
plt.show()
```

CORRELATION

```python
corr_mat=df_indexed.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_mat, cmap="crest", annot=True)
plt.title('Correlation Plot')
plt.show()
```
